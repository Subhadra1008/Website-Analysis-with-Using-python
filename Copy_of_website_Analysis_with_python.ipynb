```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/data-export (1) (1).csv")
df.head()
```

```python
df.columns=df.iloc[0]
df = df.drop(index=0).reset_index(drop=True)
df.columns = [
    "channel group",
    "Datehour",
    "Users",
    "Sessions",
    "Engaged sessions",
    "Average engagement time per session",
    "Engaged sessions per user",
    "Events per session",
    "Engagement rate",
    "Event count"
]
```

```python
df.head()
```

```python
df.info()
```

```python
df["Datehour"] = pd.to_datetime(df["Datehour"], format="%Y%m%d%H", errors="coerce")
df.head()
```

```python
numeric_cols = df.columns.drop(["channel group", "Datehour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
df["Hour"] = df["Datehour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("Datehour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and users over time")
plt.xlabel("Datehour")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title("👥 Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", y="Average engagement time per session")
plt.title("Avg Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="channel group", y="Engagement rate", palette="coolwarm")
plt.title("📦 Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
session_df = df.groupby("channel group")[["Sessions", "Engaged sessions"]].sum()

session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]

session_df_melted = session_df.melt(
    ignore_index=False,
    value_vars=["Engaged sessions", "Non-Engaged"],
    var_name="variable",
    value_name="value"
)

plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x="channel group", y="value", hue="variable")
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("⏰ Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

```python
df["Engagement rate"] = pd.to_numeric(df["Engagement rate"], errors="coerce")
df["Sessions"] = pd.to_numeric(df["Sessions"], errors="coerce")
df_plot = df.groupby("Datehour")[["Engagement rate", "Sessions"]].mean().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(df_plot["Datehour"], df_plot["Engagement rate"], label="Engagement rate")
plt.plot(df_plot["Datehour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("📊 Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
